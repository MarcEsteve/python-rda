# Cómo descargar emails de tu bandeja de entrada y parsear valores

```python
import imapclient
```

### Autenticación en Gmail

```python
conxGmail = imapclient.IMAPClient("imap.gmail.com", ssl=True)
```

```python
conxGmail.login("marc.esteve.garcia@gmail.com","lwlj sygs krqg shlb")
```

### Entramos a los datos

```python
conxGmail.select_folder("INBOX")
```

```python
#Filtrar por asunto contiene una palabra
UIDs_asunto = conxGmail.search(['SUBJECT', 'WordPress'])
```

```python
UIDs_asunto
```

```python
if UIDs_asunto:
    # Recuperar el contenido del primer mensaje encontrado
    mail_bruto = conxGmail.fetch([UIDs_asunto[0]], ['BODY[]', 'FLAGS'])
    print(mail_bruto)
else:
    print("No se encontraron mensajes con el asunto 'WordPress'.")
```

### Parsear (legible) y extraer el correo del remitente

```python
if UIDs_asunto:
    try:
        # Recuperar el contenido del primer mensaje encontrado
        mail_bruto = conxGmail.fetch([UIDs_asunto[0]], ['BODY[]', 'FLAGS'])
        mensaje = mail_bruto[UIDs_asunto[0]][b'BODY[]'].decode("utf-8")
        print(mensaje)
    except KeyError:
        print("No se pudo encontrar el cuerpo del mensaje.")
    except Exception as e:
        print(f"Ocurrió un error: {e}")
else:
    print("No se encontraron mensajes con el asunto 'WordPress'.")
```

```python
mensaje.find('From: "WordPress.com" <')
```

```python
mensaje[mensaje.find('From: "WordPress.com" <'):mensaje.find('From: "WordPress.com" <')+100]
```

```python
contiene_correo = mensaje[mensaje.find('From: "WordPress.com" <'):mensaje.find('From: "WordPress.com" <')+100]
```

```python
info_correo = contiene_correo[len('From: "WordPress.com" <'):contiene_correo.find(">")]
```

```python
info_correo
```

```python
# Cerrar la conexión
conxGmail.logout()
```

## SEARCH & FILTER Gmail

```python
con = imapclient.IMAPClient("imap.gmail.com", ssl=True)
```

```python
con.login("marc.esteve.garcia@gmail.com","lwlj sygs krqg shlb")
```

```python
con.select_folder("INBOX")
```

```python
UIDS = con.search(['SINCE', '01-Jan-2025'])
```

```python
UIDS_no_leidos = con.search(['UNSEEN'])
```

```python
UIDS_no_leidos
```

```python
for uid in UIDS_no_leidos:
    mail = con.fetch([uid], ['BODY[]', 'FLAGS'])
    print(mail)
```
